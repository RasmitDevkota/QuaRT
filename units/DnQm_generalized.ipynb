```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 2
m = 8

if n == 2:
    n_direction_groups = int(np.log2(m))-1

    direction_angles = []

    n_k_list = []
    for j in range(n_direction_groups):
        theta_0 = 90/2**j

        n_k = int(360/(theta_0))
        
        for k in range(n_k):
            theta = theta_0 + k * theta_0
            print(theta)
            direction_angles.append(theta)
    
    direction_cxs = [np.cos(theta*np.pi/180) for theta in direction_angles]
    direction_cys = [np.sin(theta*np.pi/180) for theta in direction_angles]
    direction_vectors = list(zip(direction_cxs, direction_cys))
    
    print(direction_vectors)

    fig, ax = plt.subplots(figsize=(8,8))
    ax.scatter(direction_cxs, direction_cys)
    plt.show()
```
